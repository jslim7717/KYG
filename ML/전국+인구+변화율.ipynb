```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('출생아수.xlsx')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['연도'], df['출생아 수'])
plt.title('연도별 출생아 수')
plt.xlabel('연도')
plt.ylabel('출생아 수')
```
