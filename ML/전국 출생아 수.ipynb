```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('2012 출생아 수.xlsx')
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['연도'], df['출생아 수'], color='skyblue')
plt.title('2012~2023 출생아 수')
plt.xlabel('연도')
plt.ylabel('출생아 수')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['연도'], df['출생아 수'], color='skyblue' , marker='o', linestyle='-')
# 그래프 제목 설정
plt.title('2012~2023 출생아 수')
# x축 라벨 설정
plt.xlabel('연도')
# y축 라벨 설정
plt.ylabel('출생아 수')
plt.grid(True)
plt.show()
```

```python
# 예제 데이터프레임 생성
data = {'연도': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
        '출생아 수': [484550, 436455, 435435, 438420, 406243, 357771, 326822, 302676, 272410, 249165, 214819, 190000]}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 8))

# 출생아 수 라인 그래프
plt.plot(df['연도'], df['출생아 수'], marker='o', linestyle='-', color='skyblue', label='출생아 수')

# 그래프 제목 설정
plt.title('2012~2023 출생아 수(라인 그래프)', fontsize=16)

# x축 라벨 설정
plt.xlabel('연도', fontsize=14)

# y축 라벨 설정
plt.ylabel('출생아 수', fontsize=14)

# x축 눈금 세분화
plt.xticks(df['연도'], rotation=45)

# y축 눈금 세분화
plt.yticks(range(0, 600000, 50000))

# 그리드 추가
plt.grid(True, linestyle='--', alpha=0.7)

# 데이터 레이블 추가
for i in range(len(df)):
    plt.text(df['연도'][i], df['출생아 수'][i] + 5000, f"{df['출생아 수'][i]:,}", ha='center', va='bottom', fontsize=10)

# 범례 추가
plt.legend()

plt.show()
```

```python
# 예제 데이터프레임 생성
data = {'연도': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
        '출생아 수': [484550, 436455, 435435, 438420, 406243, 357771, 326822, 302676, 272410, 249165, 214819, 190000]}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 8))

# 막대 그래프
plt.bar(df['연도'], df['출생아 수'], color='skyblue')

# 그래프 제목 설정
plt.title('2012~2023 출생아 수(히스토그램)', fontsize=16)

# x축 라벨 설정
plt.xlabel('연도', fontsize=14)

# y축 라벨 설정
plt.ylabel('출생아 수', fontsize=14)

# x축 눈금 세분화
plt.xticks(df['연도'], rotation=45)

# y축 눈금 세분화
plt.yticks(range(0, 600000, 50000))

# 그리드 추가
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 데이터 레이블 추가
for i in range(len(df)):
    plt.text(df['연도'][i], df['출생아 수'][i] + 10000, f"{df['출생아 수'][i]:,}", ha='center', va='bottom', fontsize=10)

plt.show()
```
